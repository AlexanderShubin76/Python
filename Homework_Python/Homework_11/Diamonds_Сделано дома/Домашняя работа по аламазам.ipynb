```python
# Подгружаем библиотеку pandas
import pandas as pd
```

```python
# 1. Напишите программу Pandas для чтения CSV-файла из указанного источника и печати первых 5 строк.
df.head()
```

```python
# 2. Напишите программу Pandas, чтобы выбрать серию из бриллиантов DataFrame. Распечатайте содержание серии.
df['carat']
```

```python
# 3. Напишите программу Pandas для создания новой серии «Качество? Цвет»
# (используйте обозначение в скобках для определения имени серии) бриллианта DataFrame.
df['Cut - color'] = df.cut + ', ' + df.color
df.head()
```

```python
# Напишите программу Pandas, чтобы найти детали бриллиантов, где длина> 5, ширина> 5 и глубина> 5.
df[(df['x'] >= 5) & (df['y'] >= 5) & (df['z'] >= 5)]
```
